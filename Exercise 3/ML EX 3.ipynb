### Code and Output : 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\TEJU\Downloads\telecom_customer_churn (1).csv")
```

```python
from sklearn import preprocessing
```

```python
data.head()
```

```python
data=data.drop(['Customer ID', 'City', 'Zip Code', 'Latitude', 'Longitude', 'Churn Category', 'Churn Reason'], axis = 1)
data.head()
```

```python
data.shape
```

```python
data['Avg Monthly Long Distance Charges'] = data['Avg Monthly Long Distance Charges'].fillna(0)
data['Avg Monthly GB Download'] = data['Avg Monthly GB Download'].fillna(0)
data['Multiple Lines'] = data['Multiple Lines'].fillna('No')
```

```python
cols = ['Internet Type', 'Online Security', 'Online Backup', 'Device Protection Plan', 'Premium Tech Support', 'Streaming TV', 'Streaming Movies', 'Streaming Music', 'Unlimited Data']
data[cols] = data[cols].fillna('No')
data.head()
```

```python
obj = list(data.select_dtypes(include=['object']).columns)
data[obj]
label_encoder = preprocessing.LabelEncoder() 
for col in obj:
    data[col] = label_encoder.fit_transform(data[col])
data.head()
```

```python
X = data.iloc[:,:-1].values
Y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42, max_iter=5000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Y_pred', y_pred)
print('Y_test', y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```

```python
from sklearn.metrics import f1_score
f1=f1_score(y_test,y_pred,average="macro")*100
print("F1 score of the model is {:.2f}".format(f1))
```

```python
from sklearn.metrics import recall_score
recall=recall_score(y_test,y_pred,average="macro")*100
print("Recall score of the model is {:.2f}".format(recall))
```

```python
from sklearn.metrics import precision_score
precision=precision_score(y_test,y_pred,average="macro")*100
print("Precision score of the model is {:.2f}".format(precision))
```

### Result : 
Therefore, we were successfully able to build a logistic regression model for the given dataset and compute the performance metrics of the same.
