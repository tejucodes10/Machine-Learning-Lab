```python
import pandas as pd
from sklearn import preprocessing 
```

```python
churn_df = pd.read_csv('telecom_customer_churn.csv')
churn_df.head()
```

```python
churn_df = churn_df.drop(['Customer ID', 'City', 'Zip Code', 'Latitude', 'Longitude', 'Churn Category', 'Churn Reason'], axis = 1)
churn_df.head()
```

```python
churn_df.shape
```

```python
churn_df['Avg Monthly Long Distance Charges'] = churn_df['Avg Monthly Long Distance Charges'].fillna(0)
churn_df['Avg Monthly GB Download'] = churn_df['Avg Monthly GB Download'].fillna(0)
churn_df['Multiple Lines'] = churn_df['Multiple Lines'].fillna('No')
```

```python
cols = ['Internet Type', 'Online Security', 'Online Backup', 'Device Protection Plan', 'Premium Tech Support', 'Streaming TV', 'Streaming Movies', 'Streaming Music', 'Unlimited Data']
churn_df[cols] = churn_df[cols].fillna('No')
churn_df.head()
```

```python
obj = list(churn_df.select_dtypes(include=['object']).columns)
churn_df[obj]
label_encoder = preprocessing.LabelEncoder() 
for col in obj:
    churn_df[col] = label_encoder.fit_transform(churn_df[col])
churn_df.head()
```

```python
X = churn_df.iloc[:,:-1].values
Y = churn_df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42, max_iter=5000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Y_pred', y_pred)
print('Y_test', y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```

```python
from sklearn.metrics import f1_score
f1=f1_score(y_test,y_pred,average="macro")*100
print("F1 score of the model is {:.2f}".format(f1))
```

```python
from sklearn.metrics import recall_score
recall=recall_score(y_test,y_pred,average="macro")*100
print("Recall score of the model is {:.2f}".format(recall))
```

```python
from sklearn.metrics import precision_score
precision=precision_score(y_test,y_pred,average="macro")*100
print("Precision score of the model is {:.2f}".format(precision))
```
